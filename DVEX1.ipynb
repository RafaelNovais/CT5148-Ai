```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/5D_chart_data.csv')

# Drop the first column (Unnamed: 0, assuming it's an index column)
data = data.iloc[:, 1:]

# Set the 'Field' column as the index
data.set_index('Field', inplace=True)

# Plot the data
ax = data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Annotate each part of the stacked bars with its corresponding value
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}', (x + width/2, y + height/2), ha='center', va='center')

plt.title('Comparison of Fields Across Details')
plt.xlabel('Field')
plt.ylabel('Values')
plt.show()
```

```python
#/content/drive/MyDrive/Master/CT5136 - Data Visualisation/5D_chart_data.csv
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/drive/MyDrive/Master/CT5136 - Data Visualisation/5D_chart_data.csv')  # Replace 'your_file.csv' with the actual file name

# Drop the first column (Unnamed: 0, assuming it's an index column)
data = data.iloc[:, 1:]

# Set the 'Field' column as the index
data.set_index('Field', inplace=True)

# Plot the data
ax = data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Annotate each part of the stacked bars with its corresponding value
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}', (x + width/2, y + height/2), ha='center', va='center')

# Add lines connecting each bar to the x-axis (vertical lines)
for idx, col in enumerate(data.columns):
    for bar in ax.patches[idx::len(data.columns)]:
        height = bar.get_height()
        x, _ = bar.get_xy()
        ax.plot([x + width/2, x + width/2], [0, height], color='black', linestyle='--', linewidth=1)
        ax.text(x + width/2, height + 0.1, f'{height:.1f}', ha='center', va='bottom', rotation=0)  # Adjust rotation

plt.title('Comparison of Fields Across Details')
plt.xlabel('Field')
plt.ylabel('Values')
plt.show()
```
