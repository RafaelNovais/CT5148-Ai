```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file into a DataFrame
file_path = '/content/drive/MyDrive/Colab Notebooks/Ireland.xlsx'
df = pd.read_excel(file_path)
df = df[df['City'] != 'Dublin']
provinces = df['Province'].unique()
sns.set_style("whitegrid")
palette = sns.color_palette("tab10", len(provinces))



# Scatter plot for population increase vs distance from Dublin
plt.figure(figsize=(10, 6))
for idx, province in enumerate(provinces):
    province_df = df[df['Province'] == province]
    sns.scatterplot(data=province_df, x='Distance KM', y='Population', alpha=0.5, palette=[palette[idx]])
    for _, row in province_df.iterrows():
        plt.text(row['Distance KM'], row['Population'], row['Countie'], fontsize=8)  # Annotate with county name

plt.title('Population Increase vs Distance from Dublin (Excluding Dublin)')
plt.xlabel('Distance from Dublin (KM)')
plt.ylabel('Population')
plt.legend(provinces, title='Province')
plt.show()

```

```python

# Scatter plot for population increase vs commute time from Dublin
plt.figure(figsize=(10, 6))
for idx, province in enumerate(provinces):
    province_df = df[df['Province'] == province]
    sns.scatterplot(data=province_df, x='Time Min', y='Population', alpha=0.5, palette=[palette[idx]])
    for _, row in province_df.iterrows():
        plt.text(row['Time Min'], row['Population'], row['Countie'], fontsize=8)  # Annotate with county name

plt.title('Population Increase vs Commute Time from Dublin (Excluding Dublin)')
plt.xlabel('Commute Time from Dublin (Minutes)')
plt.ylabel('Population')
plt.legend(provinces, title='Province')
plt.show()
```

```python

# Plotting Time vs Distance
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Time Min', y='Distance KM', hue='Province', palette='tab10', alpha=0.8)
plt.title('Distance from Dublin vs Commute Time (Excluding Dublin)')
plt.xlabel('Commute Time from Dublin (Minutes)')
plt.ylabel('Distance from Dublin (KM)')
plt.legend(title='Province')
plt.grid(True)
plt.show()
```
