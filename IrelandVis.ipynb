```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle

# Load the Excel file into a DataFrame
file_path = '/content/drive/MyDrive/Colab Notebooks/Book.xlsx'
df = pd.read_excel(file_path)

# Remove rows corresponding to Dublin
df = df[df['City'] != 'Dublin']

# Get unique provinces and assign colors to them
provinces = df['Province'].unique()
colors = cycle(plt.cm.tab10.colors[:len(provinces)])

# Create a dictionary to map provinces to colors
province_color_map = {prov: next(colors) for prov in provinces}


# Scatter plot for population increase vs distance from Dublin
plt.figure(figsize=(10, 6))
for province in provinces:
    province_df = df[df['Province'] == province]
    plt.scatter(province_df['Distance KM'], province_df['Population'], alpha=0.5, label=province, color=province_color_map[province])

plt.title('Population vs Distance from Dublin')
plt.xlabel('Distance from Dublin (KM)')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()



```

```python
# Scatter plot for population increase vs commute time from Dublin
plt.figure(figsize=(10, 6))
for province in provinces:
    province_df = df[df['Province'] == province]
    plt.scatter(province_df['Time Min'], province_df['Population'], alpha=0.5, label=province, color=province_color_map[province])

plt.title('Population vs Time from Dublin')
plt.xlabel('Commute Time from Dublin (Minutes)')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()
```
