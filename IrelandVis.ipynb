```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file into a DataFrame
file_path = '/content/drive/MyDrive/Colab Notebooks/Ireland.xlsx'
df = pd.read_excel(file_path)
df = df[df['City'] != 'Dublin']
provinces = df['Province'].unique()
sns.set_style("whitegrid")
palette = sns.color_palette("tab10", len(provinces))



# Scatter plot for population increase vs distance from Dublin
plt.figure(figsize=(10, 6))
for idx, province in enumerate(provinces):
    province_df = df[df['Province'] == province]
    sns.scatterplot(data=province_df, x='Distance KM', y='Population', alpha=0.5, palette=[palette[idx]])
    for _, row in province_df.iterrows():
        plt.text(row['Distance KM'], row['Population'], row['Countie'], fontsize=8)  # Annotate with county name

plt.title('Population vs Distance from Dublin')
plt.xlabel('Distance from Dublin (KM)')
plt.ylabel('Population')
plt.legend(provinces, title='Province')
plt.show()
```

```python
# Scatter plot for population increase vs commute time from Dublin
county_means = df.groupby(['Province', 'Countie']).agg({'Time Min': 'mean', 'Population': 'mean'}).reset_index()
plt.figure(figsize=(10, 6))
for idx, province in enumerate(provinces):
    province_df = county_means[county_means['Province'] == province]
    sns.scatterplot(data=province_df, x='Time Min', y='Population', alpha=0.5, palette=[palette[idx]])
    for _, row in province_df.iterrows():
        plt.text(row['Time Min'], row['Population'], row['Countie'], fontsize=8)  # Annotate with county name

plt.title('Mean Commute Time vs Population for each County')
plt.xlabel('Mean Commute Time from Dublin (Minutes)')
plt.ylabel('Population')
plt.legend(provinces, title='Province')
plt.show()
```

```python
# Plot mean commute time vs distance for each county

county_means = df.groupby(['Province', 'Countie']).agg({'Time Min': 'mean', 'Distance KM': 'mean'}).reset_index()
plt.figure(figsize=(10, 6))
for idx, province in enumerate(provinces):
    province_df = county_means[county_means['Province'] == province]
    sns.scatterplot(data=province_df, x='Time Min', y='Distance KM', alpha=0.5, palette=[palette[idx]])
    for _, row in province_df.iterrows():
        plt.text(row['Time Min'], row['Distance KM'], row['Countie'], fontsize=8)  # Annotate with county name

plt.title('Mean Commute Time vs Distance from Dublin for each County')
plt.xlabel('Mean Commute Time from Dublin (Minutes)')
plt.ylabel('Distance from Dublin (KM)')
plt.legend(provinces, title='Province')
plt.show()
```
