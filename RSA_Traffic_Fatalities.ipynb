```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RSA_Traffic_Fatalities.csv')
categories = data['Category']
years = data.columns[1:]  # Skip the first column which contains category names


driver_fatalities = data.iloc[0, 1:].values.astype(int)
passenger_fatalities = data.iloc[1, 1:].values.astype(int)
pedestrian_fatalities = data.iloc[2, 1:].values.astype(int)
cycling_fatalities = data.iloc[3, 1:].values.astype(int)
motorcyclist_fatalities = data.iloc[4, 1:].values.astype(int)
other_fatalities = data.iloc[5, 1:].values.astype(int)
plt.figure(figsize=(10, 6))

#Bar chart for all categories
plt.subplot(2, 2, 1)
plt.bar(years, driver_fatalities, label='Driver')
plt.bar(years, passenger_fatalities, bottom=driver_fatalities, label='Passenger')
plt.bar(years, pedestrian_fatalities, bottom=passenger_fatalities+driver_fatalities, label='Pedestrian')
plt.bar(years, cycling_fatalities, bottom=passenger_fatalities+driver_fatalities+pedestrian_fatalities, label='Pedal Cyclist')
plt.bar(years, motorcyclist_fatalities, bottom=passenger_fatalities+driver_fatalities+pedestrian_fatalities+cycling_fatalities, label='Motorcyclist')
plt.bar(years, other_fatalities, bottom=passenger_fatalities+driver_fatalities+pedestrian_fatalities+cycling_fatalities+motorcyclist_fatalities, label='Other')
plt.title('Fatalities by Category')
plt.xlabel('Year')
plt.ylabel('Fatalities')

#Line plot for fatalities
plt.subplot(2, 2, 2)
plt.plot(years, driver_fatalities, label='Driver')
plt.plot(years, passenger_fatalities, label='Passenger')
plt.plot(years, pedestrian_fatalities, label='Pedestrian')
plt.plot(years, cycling_fatalities, label='Pedal Cyclist')
plt.plot(years, motorcyclist_fatalities, label='Motorcyclist')
plt.plot(years, other_fatalities, label='Other')
plt.title('Fatalities Trend')
plt.xlabel('Year')
plt.ylabel('Fatalities')

# Legend
plt.legend(categories, loc='center left', bbox_to_anchor=(1, 1))
plt.tight_layout()

#Driver fatalities trend
plt.subplot(2, 2, 3)
plt.plot(years, driver_fatalities, label='Driver')
plt.title('Driver Fatalities')
plt.xlabel('Year')
plt.ylabel('Fatalities')

#Cycling fatalities
plt.subplot(2, 2, 4)
plt.bar(years, cycling_fatalities, label='Cycling')
plt.title('Cycling Fatalities')
plt.xlabel('Year')
plt.ylabel('Fatalities')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read data from CSV into a DataFrame
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RSA_Traffic_Fatalities.csv')
# Extract category names and years
categories = data['Category']
years = data.columns[1:]  # Skip the first column which contains category names

# Step 2: Extract relevant information
fatalities_data = data.iloc[:, 1:].values.astype(int)

# Step 3: Create the plot
plt.figure(figsize=(10, 6))

# Plot all categories with respective colors and value labels
ax = plt.gca()
bottom = None
for i, category in enumerate(categories):
    bars = plt.bar(years, fatalities_data[i], bottom=bottom, label=category)
    bottom = fatalities_data[i] if bottom is None else bottom + fatalities_data[i]
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.0f}', (bar.get_x() + bar.get_width() / 2, height),
                    ha='center', va='bottom', xytext=(0, 3), textcoords='offset points')

plt.title('Fatalities by Category')
plt.xlabel('Year')
plt.ylabel('Fatalities')

# Add legend for all categories with respective colors
plt.legend(categories, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()

# Step 4: Show the plot
plt.show()
```
