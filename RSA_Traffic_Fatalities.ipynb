```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RSA_Traffic_Fatalities.csv')
categories = data['Category']
years = data.columns[1:]


driver_fatalities = data.iloc[0, 1:].values.astype(int)
passenger_fatalities = data.iloc[1, 1:].values.astype(int)
pedestrian_fatalities = data.iloc[2, 1:].values.astype(int)
cycling_fatalities = data.iloc[3, 1:].values.astype(int)
motorcyclist_fatalities = data.iloc[4, 1:].values.astype(int)
other_fatalities = data.iloc[5, 1:].values.astype(int)
plt.figure(figsize=(10, 6))

# Plot 1: Bar chart for driver and pedestrian fatalities in 2022
plt.subplot(2, 2, 1)
plt.bar(years, driver_fatalities, label='Driver Fatalities')
plt.bar(years, pedestrian_fatalities, bottom=driver_fatalities, label='Pedestrian Fatalities')
plt.title('Driver and Pedestrian Fatalities')
plt.xlabel('Year')
plt.ylabel('Fatalities')
plt.legend()

# Plot 2: Line plot for fatalities trends
plt.subplot(2, 2, 2)
plt.plot(years, passenger_fatalities, label='Passenger')
plt.plot(years, cycling_fatalities, label='Pedal Cyclist')
plt.plot(years, motorcyclist_fatalities, label='Motorcyclist')
plt.plot(years, other_fatalities, label='Other')
plt.title('Fatalities Trend')
plt.xlabel('Year')
plt.ylabel('Fatalities')
plt.legend()

# Plot 3: Line plot for driver fatalities trend
plt.subplot(2, 2, 3)
plt.plot(years, driver_fatalities, label='Driver')
plt.title('Driver Fatalities Trend')
plt.xlabel('Year')
plt.ylabel('Fatalities')
plt.legend()

# Plot 4: Cycling fatalities in 2021 and 2022
plt.subplot(2, 2, 4)
plt.bar(['2021', '2022'], [cycling_fatalities[-2], cycling_fatalities[-1]])
plt.title('Cycling Fatalities in 2021 and 2022')
plt.xlabel('Year')
plt.ylabel('Fatalities')

plt.tight_layout()

# Step 4: Show the plot
plt.show()
```
